```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks")

leg = pd.read_csv('legislators.csv')
part = pd.read_csv('party-partisanship.csv')
partisans = pd.read_csv('partisans.csv')

```

```python
sns.set()
plt.plot(range(113),part['Democrat'],'b')
plt.plot(range(113),part['Republican'],'r')
plt.plot(range(113),partisans['majority'],'g')
plt.xlabel('Session of Congress')
plt.ylabel('Percentage')
plt.legend(['Democrat Party Unification', 'Republican Party Unification', 'Percent Agreement Between Parties'])
plt.show()
```
